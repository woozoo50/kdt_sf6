### 실습1. 국립중앙박물관 - 관람시간, 관람료 가져오기

```python
# 국립 중앙박물관 - 관람시간, 관람료 가져오기
import requests
from bs4 import BeautifulSoup

url = "https://www.museum.go.kr/site/main/home"
response = requests.get(url)
# print(response.text)

soup = BeautifulSoup(response.text, 'html.parser')

lis = soup.select("div.main-section-1 > ul > li")
print(lis)

for li in lis:
    print(li.text.strip())

```

### 실습 2. 전자신문 메인 기사 크롤링

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.etnews.com/20240820000076"
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

# title = soup.select_one("div.article_title")
title = soup.find("div", class_="article_title")
print(title.text)

# time = soup.select_one("div.article_header_sub > time")
time = soup.find("time", class_="date")
print(time.text , "\n")

'''
bodys = soup.select("div.article_txt > p")
# bodys = soup.findAll("div", class_="article_txt")
'''

body = soup.find('div', class_="article_txt")
body_p = body.find_all('p')
for i in body_p:
    print(i.text)
```
